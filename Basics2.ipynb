```python
# Basics methods in Strings

str = "THIS IS PYTHON"
#1) Lower Case or Upper Case
print(str.lower())
#      OR
str2 = input("Enter the String : ").lower()
print(str2)

#2) Capitalise
str3 = input("Enter the String: ").capitalize()
print(str3)

#3) Title
str4 = input("Enter a Statement : ").title()
print(str4)
```

```python
#4) Swapcase (upper case lower me or vice-versa)
str = 'Jargan'.swapcase()
print(str)
```

```python
#5) Searching and Replacing

text = 'python programming'
print(text.find("pro"))
print(text.replace("python", "Java"))
```

```python
#6) Splitting & Joining

str = "a,b,c"
s = str.split(",")
print('After Splitting', s)

#joining
result = ', '.join(s)
print(result)
```

#### Methods in Dictionary

```python
"""
keys() method is used to get all the keys of the dictionary.
values() method is used to get all the values of the dictionary.
items() method is used to get all the key-value pairs of the dictionary as a list of tuples.
get() method is used to get the value of a key. It returns None if the key is not found in the dictionary.
"""
```

```python
dictionary = {'Name':'AMIT', 'Fruit':'Apple', 'Vegetable':'Potato', 'List_items':[1,2,3,4,5], 'Number':34, 'a':34, 'float':3.14, 'Boolean':True, 'NoneType':None}
print(dictionary)
```

```python
# Keys()
print(dictionary.keys())
```

```python
# values()
print(dictionary.values())
```

```python
# items()
print(dictionary.items())
```

```python
# get()
print(dictionary.get('Name')) # Output: AMIT
print(dictionary.get('Boolean')) # Output: None
```

```python
# Shallow Copy of a Dictionary
""" Shallow copy of a dictionary creates a new dictionary object but does not create copies of the objects that the original dictionary 
references. Instead, it references the same objects. This means that if the original dictionary 
contains mutable objects (like lists or other dictionaries), changes to those objects will be 
reflected in both the original and the copied dictionary. However, if you modify the dictionary 
itself (like adding or removing key-value pairs), it will not affect the other dictionary. 
In contrast, a deep copy creates a new dictionary and recursively copies all objects, 
resulting in two completely independent dictionaries. """
```

```python
dict1 = {'Name':'AMIT', 'Age':20, 'Marks':99, 'City':'Delhi'}
dict2 = dict1
print(dict1)
print(dict2)

# Modifying dict1
dict1['Name'] = 'DEV'
print("New Dict1",dict1)
print("New Dict2",dict2) # Since dict2 is a reference to dict1, it also reflects the change made to dict1. 
# But we don't want this behaviour, we want dict2 to be independent of dict1. So we will use copy() method to create a shallow copy of dict1.

dict3 = dict1.copy()
dict1['Name'] = 'Rohit'
print("New Dict1",dict1)
print("New Dict3",dict3) # Now dict3 is independent of dict1, so changes made to dict1 will not affect dict3.
```

#### Nested Dictionary

```python
# Nested Dictionary

student_dict = {
    'Student1': {'Name':'AMIT', 'Age':20, 'Marks':99, 'City':'Delhi'},
    'Student2': {'Name':'Rohit', 'Age':22, 'Marks':95, 'City':'Mumbai'},
    'Student3': {'Name':'Suresh', 'Age':21, 'Marks':98, 'City':'Bangalore'}
}
print("Student's Data",student_dict)
```

```python
# Iterate over these Nested Dictionaries
for outer_key, outer_value in student_dict.items():
    print(f"Outer Key is :{outer_key} and Outer Value is : {outer_value}")
    for key_1, values_1 in outer_value.items():
        print(f"Inner Key is : {key_1} and Inner Value is : {values_1}")
```

#### Dictionary Comprehension

```python
# It helps us with using looping statements along with Conditional statements inside our Dictionary.
# Synatx
"""
{expression for item in iterable if condition}
expression -> key:value
"""
comprehensive = {x:x**2 for x in range(10) if x%2==0} # This will create a dictionary with keys as the squares of even numbers from 0 to 9 and values as None.
print(comprehensive)
```

#### Merged Dictionary

```python
dict1 = {'Name':'AMIT', 'Age':20, 'Marks':99, 'City':'Delhi'}
dict2 = {'Name':'Rohit', 'Age':22, 'Marks':95, 'City':'Mumbai'}

merged_dict = {**dict1, **dict2} # This will merge dict1 and dict2. If there are duplicate keys, the value from dict2 will be used.
print(merged_dict)
```

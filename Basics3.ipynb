```python
"""
name = "ramesh"
age = 10
marks = 75.5
"""
# List are not one of the data structures which are used for storing the data in a structured format where you can store the data of any data type 
# and that as it is one of the data structures so we can have access to the data as well that we are going to store into the list and we can apply 
# crud operations as well i.e create, read, update and delete operations on the list as well.

# since here to contain all diff info we have to create diff varibles
# but this issue can be solved using list
# List is mutable
# List is dynamic 
# List is heterogeneous 
```

```python
# Methods to Create Lists
#1- Square Bracket
my_list = [1,2,3,4,5.0,"hello",True]
print(my_list)

#2- Using List constructor
my_list2 = list((1,2,3,4,5,"Hello",True))
print(my_list2)

#3- List Comprehension & Range (Updating any index) (This shows list is mutable)
my_list3 = [1,2,3,4,5]
print(f"Before update list {my_list3}")
my_list3[0] = 'Hello'
print(f"After Update List {my_list3}")

# Above it was just updating 1 element 
#4- But here we'll update more than 1 element
listed = [1,2,3,4,5]
listed[0:3] = 10,20,30
print(listed)
```

```python
#5- Concatenation of 2 lists
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]
result = list1 + list2
print(result)

#6- Repeatation of List
print(list1*3)
```

```python
# Checking whether a element is present in a list or not

lst = [1,2,3,4,5,6,7,8,9]
check = int(input("Enter a number to Check: "))
if check in lst:
    print(f"{check} is Present in the List")
else:
    print(f"{check} is not Present in the List")
```

```python
# Aliasing

list1 = [1,2,3]
list2 = list1
# here we try to change any thing in list2 it will also affect list1
list2[0] = 10
print(list1, list2)

# But to overcome this problem we copy list1 and then assign list2 like
list3 = list1.copy()
list3[0] = 100
print(list1, list3)
```

```python
# List Slicing
my_list = [1,2,3,4,5,6,7,8,9]
print(my_list[0:5]) # it will print the element from index 0 to 4
print(my_list[5:]) # it will print the element from index 5 to end
print(my_list[:5]) # it will print the element from index 0 to 4
print(my_list[::2]) # it will print the element from index 0 to end
print(my_list[1::2]) # it will print the element from index 1 to end with step 2
print(my_list[::-1]) # it will print the element from end to start
```

```python
# Operations on List
#1- Append (it is used to add a new element at the end of the list)
a = [1,2,4]
a.append(5)
print(a)
```

```python
#2- Extend (it doesn't create a new list like we did in concatenation, it just simply extend the list with 2nd list's elements)
a = [1,2,3]
b = [4,5,6]
a.extend(b)
print(a)
```

```python
#3- Insert (add element at req index) Syntax: list.insert(index, value)
a = [1,2,3,4]
a.insert(0, "Hi Coders")
print(a)
```

```python
#4- Remove (It simply removes the value of ammount provided to it)
a = [1,2,3,4]
a.remove(3)
print(a)
```

```python
#5- Pop Method (here we provide the index which we want to remove)
a = [1,2,3,4]
a.pop(1) and a.pop(2)
print(a)
```

```python
#6- Clear (it just clear the whole list)
a = [1,2,3,4]
a.clear()
print(a)
```

```python
#7- Index (Used to find the index number of the element)
a = [1,2,3,4]
print(a.index(4))
```

```python
#8- Count Method (It is used to find out how many times a element appeared)
a = [1,2,3,"Python", "JAVA", True, 1,2,1,1,1]
counter = a.count(1)
print(counter)
counter2 = a.count(2)
print(counter2)
```

```python
#9- Sorting
a = [90, 50, -1, 25, 100, 0]
a.sort()
print(a)
```

```python
#10- Reverse Method
a = [1,2,3,4,5]
a.reverse()
print(a)
```

```python
#11- Copy Method (Same as we saw above)
a = [1,2,3,4,5,6]
b = a.copy()
b[0] = 100
print(a,b)
```

```python
#12- len Method (it is used to find the length of the list)
a = [1,2,3,4,5]
length = len(a)
print(length)
```

```python
#13- Enumerate Function (it is used to get the index and value of the list at the same time)
a = [11,24,15,19,10,7,3,29]
for index, value in enumerate(a):
    print(f"Index: {index}, Value: {value}")
```

#### How To Find Common Elements in 2 lists

```python
a = [1,2,3,4,5,6,7]
b = [4,5,6,7,8,9,10]
# first we convert these two lists in sets then we use set intersection method to find the common elements and then convert the new set into list again
s1 = set(a)
s2 = set(b)

s3 = s1.intersection(s2)
print(list(s3))
```

### Creating a List using List Comprehension

```python
# Syntax 
# [expression for item in interable if condition]
"""
e - x*2 (this is expression)(expression is the value which we want to store in the list)
item -  (elemnt we need in list)
iterable - range(1,11)
condition (this is optinal)
"""
# Example(Write program to get squares of all elements in range)
""" this is traditional method
squares = []
for i in range(1,10):
    squares.append(i**2)
print(squares)
"""

# Now using List Comprehension and here we put 1 condition too like we only want squares of the even numbers
squares = [i**2 for i in range(1,20) if i%2 == 0]
print(squares)
```

```python
# Example of list comprehension
words = ["Dog", "Cat", "Elephant", "Lion", "Tiger"]
length_of_words = [len(word) for word in words]
print(length_of_words)
```

```python
# Example of List Comprehension with nested list
pair = [[i,j] for i in range(1,4) for j in range(1,4)]
print(pair)
```

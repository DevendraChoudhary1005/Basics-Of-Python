## Sets

```python
# sets are going to be our 3rd data structure which can store values within the curly 
# braces format {} and are known for storing the values which are unique and unordered.
# We can also use set() function to create a set.
```

```python
a = set() # This will create an empty set.
print(type(a))
```

```python
set_value = {1, 2, 3, 4, 5} # This will create a set with the values 1, 2, 3, 4, and 5.
print(set_value)    
print(type(set_value)) 
```

#### Method in sets

```python
# 1). add() -> This method is used to add a single element to the set. If the element already exists, it will not be added again.

set_value.add(6) # This will add the element 6 to the set.
set_value.add(3) # This will not add the element 3 to the set as it already exists.
print(set_value)

# adding using iterstion
for i in range(1,25):
    set_value.add(i)
print("After adding Values : ",set_value)
```

```python
furits = {'Banana', 'Apple', 'Orange'}
vegetables = {'Tomato', 'Potato', 'Onion'}

for items in vegetables:
    furits.add(items)
print(furits)   
    
```

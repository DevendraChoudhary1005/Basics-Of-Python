```python
"""
Numeric Type:- 
int(whole numbers), 
float(decimal numbers), 
complex (real and imaginery parts)
"""
# Example int, float, complex
a = 10 
b = 10.2
c = 3+4j
print(type(a), type(b), type(c))
```

```python
"""
None Type:- Null Values
Sequesnce Type:- 
List, String, Tuple
Set Type:- Contains only unique value
set (mutable), 
frozenset (immutable means we can't modify this set)
"""
# Example for set type
unique_number = {1,2,3,3,4,5,5}
print(unique_number) # this is print a set as none the values is repeating on printing

# Example for frozenset
immutable_set = frozenset([1,2,3,3,4,4,5])
print(immutable_set)
```

### Operators


```python
# Arithematic Operators

x = 10
y = 20
print(x+y)
print(x-y)
print(x*y)
print(x/y) # remainder
print(x//y) # decimal removal (floor operator)
print(x%y) # division
print(x**y) # power
```

```python
# Comparison Operator (Boolean Output - Ture/False)

x = 10
y = 5

print(x==y)
print(x!=y)
print(x<y)
print(x>y)
print(x>=y)
print(x<=y)
```

```python
# Logical Operators (Multiple Conditions - Boolean Ture/False)
"""
and - all conditions must be true
or - atleast one condition must be true
not - reverse whatever we have
"""

age = 20
age2 = 25
is_student = True

print(age>18 and is_student)
print(age>25 and is_student)
print(age>10 and age2==25)
print(not is_student)

```

```python
# Assignment Opeartor
"""
=
+=  (x = x+5 -> x += 5) these both things means the same but 2nd is more professional
-=
*=
"""
```

```python
# Identity Opeartor
"""
It returns true if the memory location of 2 variables is same
is - returns True if the memory location of both is same 
is not - True if the memory laction is not same
"""
a = [1,2,3]
b = a
c = [1,2,3]

print(a is b) # this will provide true as a and b memory is stored at same location
print(a is c) # this will be false as a=c but not stored at same memory location
```

```python
# Membership Operator
"""
Same as Identity but here operators are "in" and "not in"
in
not in
"""

vegetables = ['Potato', 'Carrot']
print('Potato' in vegetables)
print('Lady_Finger' in vegetables)
print('LAdy_Finger' not in vegetables)
```

```python
# Simple Calculator:

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
operation = input("Enter the operation you want to perform (+, -, *, /): ")

if operation == '+':
    print(f"Sum: {a + b}")
elif operation == '-':
    print(f"Difference: {a - b}")
elif operation == '*':
    print(f"Product: {a * b}")
elif operation == '/':
    if b != 0:
        print(f"Quotient: {a / b}")
    else:
        print("Error: Division by zero is not allowed.")
else:
    print("Invalid operation. Please enter one of +, -, *, /.")
```

### Control Statements

```python
# Used for controlling the flow of the program
# Control Statements :- 
"""
conditional statements -if, elif ,else
for, while, else, suite
nested
infinite loop
pass
continue
break
assert, return
"""
```

```python
# Conditional Statements
# Example

age = int(input("Enter Your Age: "))
if(100>age>=18):
    print("You can Vote")
elif(age<0):
    print("Invalid Age")
elif(age>100):
    print("Check your age")
else:
    print("You can't Vote")
```

```python
# Example 2

num1 = int(input("Enter the 1st number: "))
num2 = int(input("Enter the 2nd Number: "))

choice = input("Enter the operator(+,-,*,/,%): ")

if(choice=='+'):
    print(num1+num2)
elif(choice=='-'):
    print(num1-num2)
elif(choice=='*'):
    print(num1*num2)
elif(choice=='/'):
    print(num1/num2)
```

```python
# Loops - while(loops ends when conditions meets truth), for(runs untill a provided limit)
# while loop syntax
""" 
while(condition):
    code execute
"""
# Example
i=int(input("Enter a Number: "))
while i<=5:
    print(i)
    i += 1
```

```python
# for loop syntax
"""
for vairable in sequence:
    # code execute
"""
# Example

for i in range(1,3):
    for j in range(1,4):
        print(f'{i},{j}')
```

```python
# break statement (used to stop code at certain condition)
# Example

for num in range(1,10,1):
    if num == 5:
        break

    print(num)
```

```python
# Continue statement(used to skip any particular step)

for num in range(1,10,1):
    if num==6:
        continue
    print(num)
```

```python
# pass statement (dummy statement)

for num in range(1,10,1):
    if num == 5:
        pass
    print(num)
```

```python
# Example on Control statements

start = int(input("Enter the start number: "))
stop = int(input("Enter the stop number: "))

skip = int(input("Enter the number you want to skip: "))

for i in range(start, stop):
    if start==stop:
        print(i)
    elif start>stop:
        break
    elif i == skip:
        continue
    else:
        print(i)
```

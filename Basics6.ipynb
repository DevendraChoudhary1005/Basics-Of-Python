## Conditional Statements

```python
# Eligibilty for driving on meeting a certain condition

age = int(input("Enter your age: "))

if age >= 18:
    print("You are eligible to drive.")
else:
    print("You are not eligible to drive.")
```

```python
# Example

a = int(input("Enter your Height:"))
b = int(input("Enter your Height:"))
c = int(input("Enter your Height:"))

if a > b and a > c:
    print("a is the tallest")
elif b > a and b > c:
    print("b is the tallest")
else:
    print("c is the tallest")
```

#### Nested Conditional Statements

```python
a = int(input("Enter a's Height:"))
b = int(input("Enter b's Height:"))
c = int(input("Enter c's Height:"))

if a > b:
    if a > c or b > c:
        print("a is the tallest")
elif b > a:
    if b > c or a > c:
        print("b is the tallest")

else:
    print("c is the tallest")
```

# Loops

#### For Loop

```python
# Loops are used for iterating over a sequence (like a list, tuple, dictionary, set, or string) or other iterable objects.
# Print the numbers from 1 to 10 
# using a for loop

for i in range(1,11):
    print(i)
```

#### Range Function


```python
# range(start, stop, step) can be used for loops to specify the starting point, stopping point, and step size for the loop iteration.
"""Parameters:
start (optional): The starting point of the sequence. Default is 0.
stop: The end point of the sequence (exclusive).
step (optional): The increment value between each number in the sequence. Default is 1.
"""
```

```python
# Using loop for decrementing the value

for i in range(10,0,-1):
    print(i) 
```

```python
# In Looping Statements, we can iterate through iterables like a string as well.

Name = 'Rajveer Singh'
for i in Name:
    print(i)
```

#### While loop

```python
# while loop is used to execute a block of code repeatedly as long as a given condition is true.
# using while loop to print numbers from 1 to 10

i = 0
while i < 10:
    i += 0.5
    print(i)
```

### Loop Control Statements

#### Loop control statements are being used for terminating a loop in between when a condition is met or if you want to have some control over your iterating process or while iterating over the series of numbers

```python
# break -> The break statement is used to exit a loop prematurely when a certain condition is met. It can be used in both for and while loops.
# Example of break statement, print no. from 1 to 10 but stop when it reaches 5

for i in range(1, 11):
    if i == 5:
        break
    print(i)
```

```python
# continue -> The continue statement is used to skip the current iteration of a loop and move on to the next iteration. It can also be used in both for and while loops.
# Example of continue statement, print no. from 1 to 10 but skip when it reaches 5

for i in range(11):
    if i == 5:
        continue
    print(i)
    
```

```python
# pass -> The pass statement is a null operation; it does nothing when executed. It is used as a placeholder in situations where syntactically some code is required but you don't want to execute any code. It can be used in loops, functions, classes, or any block of code.
# Example of pass statement, using pass in a loop

for i in range(5):
    if i == 2:
        pass  # This will do nothing and the loop will continue
    print(i) 
```

### Nested Loops

```python
for i in range(1, 5):
    for j in range(1, 4):
        print(i,j)
```

```python
# Examples -> Print the sum of first n Natural Numbers

n = int(input("Enter the Values of n: "))
sum = 0 
for i in range(1, n+1):
    sum += i
print(f"The sum of first {n} Natural Numbers is: {sum}")
```

```python
# Example -> Check whether the given number is prime or not

n = int(input("Enter a number: "))

for i in range(2, n):
    is_prime = True
    for j in range(2, i):
        if i % j == 0:
            is_prime = False
            break
    if is_prime:
        print(f"{i} is a prime number")
```
